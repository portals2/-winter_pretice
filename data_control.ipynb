```python
import string

src_str = string.ascii_uppercase
dst_str = src_str[3:] + src_str[:3]
re_str = dst_str[-6:] + dst_str[:-6]

def cipher(a):
    idx = src_str.index(a) #A~Z까지의 인덱스 구하기
    return dst_str[idx]

def Return(r):
    idx = src_str.index(r)
    return re_str[idx]
    

src = input('문자')
print('암호문 :',  end='')

for ch in src:
    if ch in src_str:
        k = cipher(ch)
        print(k, end='')
    else:
        print(ch, end='')

print('복구k :',  end='')
for ch in src:
    if ch in src_str:
        k = cipher(ch)
        print(Return(k), end='')
    else:
        print(ch, end='')
```
