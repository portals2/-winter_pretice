```python
import string

src_str = string.ascii_uppercase
dst_str = src_str[3:] + src_str[:3]
re_str = dst_str[-6:] + dst_str[:-6]

def cipher(a):
    idx = src_str.index(a) #A~Z까지의 인덱스 구하기
    return dst_str[idx]

def Return(r):
    idx = src_str.index(r)
    return re_str[idx]
    

src = input('문자')
print('암호문 :',  end='')

for ch in src:
    if ch in src_str:
        k = cipher(ch)
        print(k, end='')
    else:
        print(ch, end='')

print('복구k :',  end='')
for ch in src:
    if ch in src_str:
        k = cipher(ch)
        print(Return(k), end='')
    else:
        print(ch, end='')
```

```python
import numpy as np
from PIL import Image
import wikipedia as wp
from wordcloud import *
import matplotlib.pyplot as plt

wiki = wp.page('Aircraft')
txt = wiki.content

s_words = STOPWORDS.union({'km', 'h', 'm', 's', 'may'})
world_mask = np.array(Image.open("airplane.png"))
wdcloud = WordCloud(background_color='white', max_words=1000, stopwords=s_words, mask=world_mask, colormap='OrRd').generate(txt)

plt.figure(figsize=(120,120))
plt.imshow(wdcloud, interpolation='bilinear')
plt.axis('off')
plt.savefig('test1.png', dpi = 300)
plt.show()
```
