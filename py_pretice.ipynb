```python
#함수의 사용 및 재사용
#터틀을 이용한 마크 찍기
import turtle as t
import random


t.shape("circle")
t.colormode(255)
t.hideturtle()
t.speed(0)

def rgb():
    r = random.randrange(0, 256)
    g = random.randrange(0, 256)
    b = random.randrange(0, 256)
    return r, g, b

def move(x, y):
    t.penup()
    t.goto(x,y)
    t.pendown()
    
def draw_heart(x, y):
    move(x, y)
    t.pencolor(rgb())
    t.fillcolor(rgb())
    t.begin_fill()
    t.left(45)
    t.forward(40)
    t.circle(15, 221.37)
    t.left(180)
    t.circle(15, 221.37)
    t.forward(40)
    t.end_fill()
    
while(True):
    draw_heart(random.randrange(-750, 750), random.randrange(-450, 450))

# while(True):
#     t.color(rgb())
#     t.goto(random.randrange(-100, 100), random.randrange(-100, 100))
#     t.stamp()

t.done()
```

```python
import turtle

def bar(h):
    t.begin_fill()
    t.left(90)
    t.forward(h)
    t.write(str(h), font=("Times New Roman", 16, "bold"))
    t.right(90)
    t.forward(40)
    t.right(90)
    t.forward(h)
    t.left(90)
    t.end_fill()
    
data = [120, 56, 309]

t = turtle.Turtle()
t.color("blue")
t.fillcolor("red")
t.pensize(3)

for d in data:
    bar(d)


turtle.done
```
