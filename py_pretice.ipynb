```python
import random

tries = 0
guess = 0
answer = random.randint(1, 100)
print("guess", answer)
while guess != answer:
    guess = int(input("num"))
    tries += 1
    if guess < answer:
        print("low")
    elif guess > answer:
        print("hi")
    else:
        print("con, try=", tries)
    if tries == 10:
        break
```
