```python
c = [('s', 1), ('b', 2), ('i', 3)]
p_=0
for i in range(len(c)):
    max_c = max(i[1] for i in c) #0[1], 1[2]... 순으로 바뀌고 비교
    min_c = min(i[1] for i in c)
    p_ += c[i][1]

print(max_c)
print(min_c)
print(p_ / len(c))
```
