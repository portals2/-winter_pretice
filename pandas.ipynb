```python
import datetime as dt

weather = pd.read_csv('ww.csv', encoding='CP949')

monthly = [None for x in range(12)]
monthly_temp = [0 for x in range(12)]
weather['day'] = pd.DatetimeIndex(weather['날짜']).month
for i in range(12):
    monthly[i] = weather[weather['day'] == i + 1]
    monthly_temp[i] = monthly[i].mean()['평균기온(℃)']
    
plt.plot(monthly_temp, 'red')
plt.show()

weather['평균기온(℃)'].plot()
weather['최저기온(℃)'].plot()
weather['최고기온(℃)'].plot()
plt.show()

weather['month'] = pd.DatetimeIndex(weather['날짜']).month
means = weather.groupby('month').mean()
print(means)

means['평균기온(℃)'].plot(kind='bar')

plt.show()
```

```python
df_1 = pd.DataFrame({'item': ['1', '1', '3', '3'],
                    'type' : ['10', '02', '03', '04'],
                    'price' : [20000, 10000, 50000, 30000]})

print(df_1)
df_2 = df_1.pivot(index="item", columns='type', values='price') #pivot으로 행렬 바꿈
print(df_2)
```

```python
df_1 = pd.DataFrame({
    'A': ['a10', 'a11', 'a12'],
    'B' :['b10', 'b11', 'b12'],
    'C' :['c10', 'c11', 'c12']
}, index=['가', '나', '다'])
df_2 = pd.DataFrame({
    'B': ['b10', 'b11', 'b12'],
    'C' :['c10', 'c11', 'c12'],
    'D' :['d10', 'd11', 'd12']
}, index=['다', '라', '마'])

df_3 = pd.concat([df_1, df_2])
print(df_3, 'aaaaaaaa')

#axis 0은 가로 1은 세로, join outer는 데이터 합치기, inner은 중복 데이터 출력
print(pd.concat([df_1, df_2], axis=0, join='outer'), 'ccccccccc')
print(pd.concat([df_1, df_2], axis=0, join='inner'), 'ccccccccc')
print(pd.concat([df_1, df_2], axis=1, join='outer'), 'ccccccccc')
print(pd.concat([df_1, df_2], axis=1, join='inner'), 'ccccccccc')
```

```python
#inplace는 안쪽에 자료가 존재할 때 적용
countries_df = pd.read_csv('e.csv', index_col = 0)
sorted =countries_df.sort_values('pop')
countries_df.sort_values('pop', inplace=True)
print(sorted)
print(countries_df)

#ascending = False >> 내림차순 // 기본은 오름차순
countries_df.sort_values(['area'],ascending=False, inplace=True) 
print(countries_df)
```
