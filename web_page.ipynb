```python
from bs4 import BeautifulSoup

import requests
import numpy as np
import time as t

req = requests.get('https://www.dhlottery.co.kr/gameResult.do?method=statByNumber')
html = req.text

soup = BeautifulSoup(html, 'html.parser')
result = soup.find_all('tbody')
result1 = soup.find_all('td')[0] #4 >> +3 1:4 2:7 3:10 ... 136
per = []
for i in range(4, 137, 3):
    per.append(int(soup.find_all('td')[i].get_text()))
z = []
for i in range(len(per)):
    z.append(per[i] / sum(per))


print(sum(z))

a = np.arange(1,46)
for i in range(5):
    b = np.random.choice(a, 6, replace= False, p = z)
    print(np.sort(b))
#결과물 웹사이트에 올리기
```
