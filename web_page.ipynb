```python
import urllib

URL = 'https://www.naver.com'
response = urllib.request.urlopen(URL)
byte_data = response.read()
text_data = byte_data.decode('utf-8')
print(text_data)
```

```python
import urllib as ul

parse1 = ul.parse.urlparse('https://www.youtube.com/watch?v=A-30TFlnyGs')
parse2 = ul.parse.urlsplit('https://www.youtube.com/watch?v=A-30TFlnyGs')
print(parse1)
print(parse2)
qs = ul.parse.parse_qs(parse1.query)
print(qs)
qs1 = ul.parse.parse_qsl(parse1.query)
print(qs1)
re = ul.parse.urlunparse(parse1)
print(re)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup('<script></script>', 'lxml')
print(soup)
```

```python
req = requests.get('https://www.naver.com')
html = req.text
soup = BeautifulSoup(html, 'html.parser')

print(soup.title)
print(soup.title.name)
print(soup.title.string)

print(soup.find('a'))
print(soup.find_all('a')[1])

print(soup.select_one('body a a'))

text = soup.find('span', attrs={'class' : 'blind'})
print(text.get_text())
```
